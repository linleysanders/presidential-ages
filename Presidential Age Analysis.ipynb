# An Analysis of American Presidents By Age

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("presidents_age.csv", encoding="utf8")
````

````python
df.shape
````

````python
df.dtypes
````

````python
#Don't panic if the index number appears off. We didn't drop a president. George Washington starts at 0.
df.head()
````

## So, the ```Age at Death``` column isn't incredibly helpful for an analysis, since we're throwing strings and integers together. Also, who gives their age in terms of days?

````python
df['Year'] = df['Age at Death'].str.extract(r'(^\d\d)').astype(int)
````

## Now we have the basic age of each American president in the ```Year``` column

````python
df.tail(10)
````

````python
#The average age for a president is around 71 years old.
df['Year'].mean()
````

````python
age = df['Year']
````

````python
ax = age.plot(kind='hist')

plt.rcdefaults()
plt.style.use('fivethirtyeight')

ax.set_title("American Presidents By Age")
ax.set_ylabel("Frequency")
ax.set_xlabel("U.S. Presidential Ages (Living and Deceased)")

plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42
})

plt.savefig("president-hist.pdf")
plt.show()
````

````python
ax = age.plot(kind='hist', bins=7)

plt.rcdefaults()
plt.style.use('fivethirtyeight')

ax.set_title("American Presidents By Age")
ax.set_ylabel("Frequency")
ax.set_xlabel("U.S. Presidential Ages (Living and Deceased)")

plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42
})

plt.show()
````

````python
print("Mean Age: ", df['Year'].mean())
print("Standard deviation:", df['Year'].std())
print("Minimum Age: ", df['Year'].min())
print("Maximum Age: ", df['Year'].max())
````

````python
print("25th percentile:   ", np.percentile(df['Year'], 25))
print("Median:            ", np.median(df['Year']))
print("50th percentile:   ", np.percentile(df['Year'], 50))
print("75th percentile:   ", np.percentile(df['Year'], 75))
print("85th percentile:   ", np.percentile(df['Year'], 80))
print("90th percentile:   ", np.percentile(df['Year'], 90))
print("99th percentile:   ", np.percentile(df['Year'], 99))
````
